Q1: 
    + My data set is "Zoo data set" (link: https://archive.ics.uci.edu/ml/datasets/Zoo )
    + Zoo data set has total 18 attributes:
        1. animal name: Unique for each instance
        2. hair: Boolean : have hair or not
        3. feathers: Boolean : have feathers or not
        4. eggs: Boolean : lay eggs or not
        5. milk: Boolean : have milk or not
        6. airborne: Boolean : can fly or not
        7. aquatic: Boolean  : live underwater or not
        8. predator: Boolean : kill, eat other animal or not
        9. toothed: Boolean : have tooth or not
        10. backbone: Boolean : have backbone or not
        11. breathes: Boolean : can breathe or not
        12. venomous: Boolean : have poison or not
        13. fins: Boolean :  have fins or not
        14. legs: Numeric (set of values: {0,2,4,5,6,8}) : amount of legs
        15. tail: Boolean : have tail or not
        16. domestic: Boolean : an animal that is not wild and is kept as a pet or to produce food or not
        17. catsize: Boolean
        18. type: Numeric (integer values in range [1,7]) : 7 type animals
    + Model's goal is depend to first 17 attributes to decide the type of each animal from 1 to 7 in the Zoo
    + Describe data graphically:

```python
# Read data set by pandas
import pandas as pd
import numpy as np

data = pd.read_csv("zoo.data", header=None).values
print(data)
print("--------------------")
#transformers, and normalizers
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
good_data= scaler.fit_transform(data[:,1:-1])

data[:,1:-1]=good_data

print(data,"data good")
```

```python

import matplotlib.pyplot as plt
#statistics the number of each type
countType =np.zeros(7)
mylabels = ["Type 1", "Type 2", "Type 3", "Type 4", "Type 5", "Type 6", "Type 7"]
for row in data:
    countType[row[-1]-1]+=1

plt.pie(countType, labels = mylabels, startangle = 90,autopct='%1.1f%%',)
plt.show() 


```

```python
# Try to analysis the predator animal and animal have eggs of each type
# Because, most animals have eggs usually do not kill, eat other animals and vice versa.
amountPredator=np.zeros(7)
amountAEggs=np.zeros(7)
for row in data:
    if row[3]==True:
        amountAEggs[row[-1]-1]+=1
    if row[7]==True:
        amountPredator[row[-1]-1]+=1
x = np.arange(len(mylabels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, amountPredator, width, label='Predator')
rects2 = ax.bar(x + width/2, amountAEggs, width, label='Lay eggs')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number')
ax.set_title('Number predator animal and animal lay eggs')
ax.set_xticks(x)
ax.set_xticklabels(mylabels)
ax.legend()


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

=> So, in type 1: the number of predator/eggs is 22/1 and type 1 have 40.6% of total animal in the zoo.
In other type, we can see the number animal have eggs slightly more than predator animals.

```python
#Next to analysis airborne and aquatic animals.
amountAirborne=np.zeros(7)
amountAquatic=np.zeros(7)
for row in data:
    if row[5]==True:
        amountAirborne[row[-1]-1]+=1
    if row[6]==True:
        amountAquatic[row[-1]-1]+=1


plt.plot(mylabels, amountAirborne, label = "Airborne" ,marker='o')
plt.plot(mylabels, amountAquatic, label = "Aquatic",marker='o')
plt.legend()
ax.set_ylabel('Number')
plt.grid(True)
plt.title('Number Airborne animal and Aquatic animal')
plt.show()

```

So, we can see that Airborne and Aquatic animals was decreased from type 1 to 7 and the amount of two animal are very small.

```python
# Question 2: Build kNN, Naive Bayes models for the task. Show model evaluation results on the test data.
# Convert column type to string to apply models
for row in data:
    row[-1]="Type "+ str(row[-1])
    
# Generate data test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data[:,:-1],data[:,-1],test_size=0.4)
```

```python
#KNN:
from sklearn.neighbors import KNeighborsClassifier


neight= KNeighborsClassifier(n_neighbors=3)

neight.fit(x_train[:,1:],y_train)
print("*Score predict with this data train" ,neight.score(x_test[:,1:],y_test))

predit_result = neight.predict(x_test[:,1:])
for i in range(len(predit_result)):
    print("Name:",x_train[i,0],"------Actual:",y_test[i], "------Predict:",predit_result[i])
```

```python
#Evaluate resule KNN model 
from sklearn.metrics import classification_report
print(classification_report(y_test,predit_result))
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,predit_result))
```

```python
#Naive Bayes:
from sklearn.naive_bayes import CategoricalNB


modelNB = CategoricalNB()
modelNB.fit(x_train[:,1:],y_train)


print("*Score predict with this data train" ,modelNB.score(x_test[:,1:],y_test))

predit_resultNB = modelNB.predict(x_test[:,1:])
for i in range(len(predit_resultNB)):
    print("Name:",x_train[i,0],"------Actual:",y_test[i], "------Predict:",predit_resultNB[i])
```

```python
#Evaluate resule Naive Bayes model 

print(classification_report(y_test,predit_resultNB))

print(confusion_matrix(y_test,predit_resultNB))
```

# Question 4: Study the Overfiting issue in machine learning and how to resolve it in programming. This question should be reported in both slides and codes.

```python
#Convert to int to use LASO , RIDGE ALGORITHMS
for i in range(len(y_train)):
    y_train[i]=int( y_train[i].split()[-1])
for i in range(len(y_test)):
    y_test[i]=int( y_test[i].split()[-1])
```

```python
#LASO
from sklearn import linear_model
laso = linear_model.Lasso(alpha=0.1) #Regularization strength
laso.fit(x_train[:,1:],y_train)
print(laso.coef_,"--Parameter vector")
print(laso.intercept_,"--Independent term in decision function.")
print("*coefficient of determination R^2 in data train" ,laso.score(x_train[:,1:],y_train))
print("*coefficient of determination R^2 in data test" ,laso.score(x_test[:,1:],y_test))
```

```python
#RIDGE
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.1) #Regularization strength
ridge.fit(x_train[:,1:],y_train)
print(ridge.coef_,"--Parameter vector")
print(ridge.intercept_,"--Independent term in decision function.")
print("*coefficient of determination R^2 in data train" ,ridge.score(x_train[:,1:],y_train))
print("*coefficient of determination R^2 in data test" ,ridge.score(x_test[:,1:],y_test))
```
